```python
import pandas as pd
import os
import random
```

```python
ORIGINAL_DATASET = "../data/all_reviews/all_reviews.csv"
PROCESSED_DATASET_PATH = "../data/all_reviews/processed_dataset.csv"
CHUNK_SIZE = 1_000_000
```

```python
columns = [
    'recommendationid', 'appid', 'game', 'author_steamid', 
    'author_playtime_forever', 'author_playtime_last_two_weeks', 
    'author_playtime_at_review', 'author_last_played', 'voted_up'
]
```

```python
if os.path.exists("PROCESSED_DATASET_PATH"):
    os.remove("PROCESSED_DATASET_PATH")

chunk_iteration = pd.read_csv(ORIGINAL_DATASET, chunksize=CHUNK_SIZE, usecols=columns)

total_rows = 0

for i, chunk in enumerate(chunk_iteration):
    
    chunk['author_playtime_forever'] = chunk['author_playtime_forever'] / 60

    chunk = chunk[chunk['author_playtime_forever'] > 2.0]

    chunk.to_csv(PROCESSED_DATASET_PATH, mode='a', index=False, header=(i==0))

    total_rows += len(chunk)

    print(f"Proccessed chunk {i+1} - Total rows saved: {total_rows}")

    if i == 65:
        print(f"{i} chunks have been processed, exiting...")
        break
```
